```python
import json
import pandas as pd
import requests
```

```python
#loads a JSON file
url = 'https://raw.githubusercontent.com/jasonwnc/ds2002s24/main/2.%20Python/coll%20bb_SortByConf.json'
response = requests.get(url)
teams = json.loads(response.text)
#put it into data frame
df = pd.DataFrame(teams['teams'])
#output 100 rows
df.head(100)
```

```python
#how many teams are there
num_teams = len(df)
print("Number of teams:", num_teams)
```

```python
#how many teams are in the state of Virginia
virginia_teams = df[df['state'] == 'VA']
print("Number of teams in Virginia:", virginia_teams)
```

```python
#list of duplicate mascots and group them by number
duplicates = df.groupby('name').filter(lambda x: len(x) > 1)['name'].value_counts()
print(duplicates)
```
